# Principal Component Analysis (PCA) using Scikit-Learn

````python
import sklearn
sklearn.__version__
````

````python
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline
````

Principal Component Analysis (PCA) reduces the dimension of the data by representing it using a pre-specified number of new (linear and orthonormal) features. Each of these features is a  directions in the input space in which the data has maximal variance. Equivalently, PCA can be viewed as a procedure for finding a low dimensional representation of the data, such that if we optimally restore the original data from this representation, then the sum-of-squred errors between original and restored data is minimized.

PCA has many usages, here we will see how it can be used to analyze images of faces.

### First, we load the data

````python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=40)
````

````python
print('The dimensions of the input data are: ' + str(faces.images.shape))
print('The labels in the data are: ' + str(list(faces.target_names)))
````

Using images for our experiments help us visualize the data, and later also the effect of our algorithms. 

````python
fig, axes = plt.subplots(3, 8, figsize=(18, 7), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.3, wspace=0.1))
for n, ax in enumerate(axes.flat):
    ax.imshow(faces.images[n], cmap='gray')
    ax.set_title(faces.target_names[faces.target[n]],fontsize=10)
````

### Finding the leading principal components

````python
from sklearn.decomposition import PCA
pca = PCA(n_components=200)
````

````python
pca.fit(faces.data);
````

````python
fig, axes = plt.subplots(3, 8, figsize=(18, 7), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.3, wspace=0.1))
for n, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[n].reshape(62, 47), cmap='gray')
    ax.set_title(str(np.round(100*pca.explained_variance_ratio_[n],1)) + '% of var ',fontsize=10)
fig.suptitle('Leading principal components, and the percetange of total variance captured by each component');
````

The results give us some insight into how the images in our data vary. For example, the few leading components appear to be associated with the direction of lighting, followed by components that appear to be associated with facial parts such as eyes and mouth.

Note in the figure below how more than 90 % of the total variance in the data is captured by it's leading 100 principal components. 

````python
plt.subplots(2,1,figsize=[15,8],sharex=True)
plt.subplot(2,1,1)
plt.plot(pca.explained_variance_ratio_)
plt.ylabel('explained variance');
plt.subplot(2,1,2)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of principal components')
plt.ylabel('cumulative explained variance');
````

The graphs above suggest that we can recover most of the information in the data using only its projection onto the leading 100-200 principal components. We can test this by projecting the compact representation back into the input space. Scikit-Learn's PCA provides this functionality. 

````python
Ncomponents = 100
Nfaces = 10
fig, ax = plt.subplots(2, Nfaces, figsize=(15, 4),subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.1, wspace=0.1))

for n in range(Nfaces):
    ax[0, n].imshow(faces.data[n].reshape(62, 47), cmap='gray')
ax[0, 0].set_ylabel('input data')

# Compute the leading principal components
pca = PCA(Ncomponents).fit(faces.data)
components = pca.transform(faces.data)

# Compute the best restoration of the original images
projected = pca.inverse_transform(components)

for i in range(Nfaces):
    ax[1, i].imshow(projected[i].reshape(62, 47), cmap='gray')
ax[1, 0].set_ylabel('Reconstruction \nusing ' + str(Ncomponents) +' PCs');
````
